### plot( )로 그래프를 그린다

```python
np.arange(0.0, 2.0, 0.01)
len(np.arange(0.0, 2.0, 0.01))
```

```python
np.arange(0.0, 2.0, 0.01).shape
```

```python
# sin그래프를 그린다
s = np.sin(np.pi*np.arange(0.0, 2.0, 0.01)) # 약 0~2pi까지 200등분 된 값 (x축 지정 안함)
plt.plot(s)    # sin의 그래프를 그림
```

### x-축을 지정하여 그리기.

```python
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.1)) # 20개의 데이터 포인트
plt.plot(np.arange(0.0, 2.0, 0.1), s1, '*', ms = 8) # marker '*'로 변경
```

```python
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.01)) # 200개의 데이터 포인트
plt.plot(np.arange(0.0, 2.0, 0.01), s1, 'o',ms = 1) # 마크 사이즈 1로 변경
```

## 산포도 (scatter plot)

```python
# 산포도(scatter plot)를 그린다
# 마커를 사용하여 산포도를 작성
# 표준 정규 분포에 따라 난수 생성

x = np.random.randn(5000)
y = np.random.randn(5000)
plt.figure(figsize=(5,5)) # 그래프 크기
t = plt.plot(x, y, 'o', alpha=0.1)   #, ms=8)
```

### [도전] 균일분포(백색잡음)의 산포도 그리기

```python
# 균일 분포에 따라 난수 생성하고 산포도 그리기

x = np.random.rand(5000)
y = np.random.rand(5000)
plt.figure(figsize=(5,5))
t = plt.plot(x, y, 'o', alpha=0.9)   #, ms=8)
```

### 막대 그래프 (Bar graph)

```python
# 막대그래프(bar graph)를 그린다.

means = [1, 2, 3] # 평균
stddevs = [0.2, 0.4, 0.5] # 표준편차
bar_labels = ['bar 1', 'bar 2', 'bar 3']

# plot bars
x_pos = list(range(len(bar_labels)))  # bar의 위치 설정
plt.bar(x_pos, means, yerr=stddevs, color=('r','g','b'))

plt.show()
```

### 히스토그램 (histogram)

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) 

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x, bins=bins)
plt.show()
```

```python
# 중첩 히스토그램
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000) 
x2 = rng.normal(20, 10, 1000)

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x1, bins=bins, alpha=0.5)
plt.hist(x2, bins=bins, alpha=0.5)
plt.show()
```
